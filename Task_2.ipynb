## 1. Create a function

```python
import matplotlib.pyplot as plt

def analyze_dna_file (file_name): # Creating the function

    # Everything below is the same code as for task 1
    with open(file_name, "r") as file:
        dna1 = file.read()
        dna1 = dna1.upper()
        sequences = dna1.split("\n") 


    for i in range(0, len(sequences),2):
        SQ_nr = sequences[i]
        dna_string = sequences [i+1]

        letter_count_SQ = {'A': 0, 'T': 0, 'C':0, 'G': 0, 'N':0}
        for letter in dna_string:
            letter_count_SQ[letter] += 1

        print(f'{SQ_nr}: {letter_count_SQ}')


        # Create bar chart
        plt.bar(letter_count_SQ.keys(), letter_count_SQ.values(), color = '#9391ff') 
        plt.title(SQ_nr) # Use sequence ID as title
        plt.xlabel('DNA Letters')
        plt.ylabel('Count')
        plt.show()


# Calling the function
analyze_dna_file("dna_raw.txt")
```

```python
import matplotlib.pyplot as plt

def analyze_dna_file (file_name): # Creating the function

   
    with open(file_name, "r") as file:
        dna1 = file.read()
        dna1 = dna1.upper()
        sequences = dna1.split(">")[1:]
        # we're using '>' as the separator since thats what every new sequence line starts with
        # to avoid getting an empty list we slice the list with [1:] which removes the firt empty element

    for sequence in sequences:
        lines = sequence.split("\n") # to separate the sequence id from the dna sequence we're spliting the new lines by newline (\n)
        SQ_id = lines[0] # the first line is the sequence id --> SEQ1, SEQ2 etc
        dna_lines = lines[1:] # Everything after the first line is the dna sequence

        dna_sequence = "".join(dna_lines) 
        # using join method to merge all DNA lines into ONE string
        # the "" menas that we join the lines without any  spaces and commas (no separators)
       

        # Same as for task 1
        letter_count_SQ = {'A': 0, 'T': 0, 'C':0, 'G': 0, 'N':0} 
        for letter in dna_sequence:
            letter_count_SQ[letter] += 1
        print(f'{SQ_id}: {letter_count_SQ}')


         # Create bar chart --> exactly the same as for task1
        plt.bar(letter_count_SQ.keys(), letter_count_SQ.values(), color = '#9391ff') 
        plt.title(SQ_id)
        plt.xlabel('DNA Letters')
        plt.ylabel('Count')
        plt.show()
   
analyze_dna_file("dna_raw_complicated.txt")
```
